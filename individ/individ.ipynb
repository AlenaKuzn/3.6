Постройте 3D-график функции f(x, y) = sin($x^2 $ + $ y^2 $) на интервале 0, 5 x 0, 5. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# создание сетки значений x и y
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
x, y = np.meshgrid(x, y)

# вычисление значения функции f(x, y)
z = np.sin(x ** 2 + y ** 2)

# построение 3D-графика функции f(x, y)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, y, z, cmap='coolwarm')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('График функции f(x, y) = sin(x^2 + y^2)')
plt.show()
```
